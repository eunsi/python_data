# matplotlib 그래프

```python
import matplotlib.pyplot as plt
```

```python
#기본 차트 그리기 x,y값 지정

x=[1,2,3,4,5]
y=[2,5,3,1,2]

plt.plot(x,y)
plt.show()
```

```python
# 딕셔너리 형태로 값 저장하고 불러오기
#중간에 있는 ''는 data= 를 지정하고, key나 칼럼 이름을 따옴표 지정하게 되면,
#다른 옵션과 혼란의 여지가 있으므로, 넣어줌.

dict1 = {'v1':[1,2,3,4,5], 'v2':[2,5,3,1,2]}

plt.plot('v1','v2','',data = dict1)
plt.show()
```

```python
# x,y축 레이블 , 타이틀 붙이기

dict1 = {'v1':[1,2,3,4,5], 'v2':[2,5,3,1,2]}
plt.plot('v1','v2','',data=dict1)
#plt.xlabel('month')
#plt.ylabel('sales')
plt.title('Monthly Sales')
plt.show()
```

```python
# 라인 스타일 조정하기
#[color],[marker],[linestyle]

dict1 = {'v1':[1,2,3,4,5], 'v2':[2,5,3,1,2]}
plt.plot('v1','v2','go--',data=dict1)

plt.title('Monthly Sales')
plt.show()
```

```python
#여러 그래프 겹쳐서 그리기
#그리드 추가 plt.grid()

dict2= {'v1':[1,2,3,4,5], 'v2':[2,5,3,1,2], 'v3':[4,6,7,6,5]}
plt.plot('v1','v2','go--',data = dict2)
plt.plot('v1','v3','ro-',data = dict2)

plt.title('Monthly Sales')
plt.show()
```

```python
#범례 추가

dict2= {'v1':[1,2,3,4,5], 'v2':[2,5,3,1,2], 'v3':[4,6,7,6,5]}
plt.plot('v1','v2','go--',data = dict2, label='a')
plt.plot('v1','v3','ro-',data = dict2 , label='b')
plt.legend()
plt.grid()
plt.title('Monthly Sales')
plt.show()
```

```python
# 축 범위 추가
# xlim, ylim
#그래프 크기 조정 (figure=(4,3))

dict2= {'v1':[1,2,3,4,5], 'v2':[2,5,3,1,2], 'v3':[4,6,7,6,5]}
plt.plot('v1','v3','ro-',data = dict2)

plt.title('Monthly Sales')
plt.xlim(0,6)
plt.ylim(0,8)
plt.grid()
plt.show()
```

```python
#수평선, 수직선 추가하기 (평균 나타낼때 사용됨)
# 수평선 추가 plt.axhline(y값, color, linestyle)
# 수직선 추가 plt.avhline(x값, color, linestyle)

dict2= {'v1':[1,2,3,4,5], 'v2':[2,5,3,1,2], 'v3':[4,6,7,6,5]}
plt.plot('v1','v3','ro-',data = dict2)

plt.title('Monthly Sales')
plt.axhline(5,color='grey',linestyle='--')
plt.axvline(2,color='grey',linestyle='--')

plt.show()
```

```python
#그래프에 텍스트 추가
# plt.text(x좌표, y좌표, '추가할 텍스트')

dict2= {'v1':[1,2,3,4,5], 'v2':[2,5,3,1,2], 'v3':[4,6,7,6,5]}
plt.plot('v1','v3','ro-',data = dict2)

plt.title('Monthly Sales')
plt.axhline(5,color='grey',linestyle='--')
plt.axvline(2,color='grey',linestyle='--')

plt.text(1.3,5.4,'5.0')

plt.show()
```

```python
# 그래프 나눠서 그리기 
# plt.subplot(rows, cols , index)
#(행, 열, 인덱스)

dict2= {'v1':[1,2,3,4,5], 'v2':[2,5,3,1,2], 'v3':[4,6,7,6,5]}

plt.subplot(2,1,1)
plt.plot('v1','v2','go--',data = dict2)
plt.subplot(2,1,2)
plt.plot('v1','v3','ro-',data = dict2)

plt.show()
```
