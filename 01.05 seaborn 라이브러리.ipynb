## seaborn 라이브러리

```python
# 라이브러리 불러오기

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
titanic = pd.read_csv('https://bit.ly/3FsgwkJ')
titanic.head()
```

```python
iris = pd.read_csv('https://bit.ly/3JiY7ZZ')
iris.head()
```

```python
boston = pd.read_csv('https://bit.ly/3EuWvZw')
boston.head()
```

```python
air = pd.read_csv('https://bit.ly/3qmthqZ')
air['Date'] = pd.to_datetime(air['Date'])
air['Month'] = air.Date.dt.month
air['Weekday'] = air.Date.dt.weekday
air.head()
```

```python
stock = pd.read_csv('https://raw.githubusercontent.com/DA4BAM/dataset/master/KOSPI.csv')
stock['Date'] = pd.to_datetime(stock['Date'])
stock['Year'] = stock.Date.dt.year
stock['Month'] = stock.Date.dt.month
stock['Weekday'] = stock.Date.dt.weekday
stock.head()
```

## seaborn 차트 그리기

```python
# histogram : sns.histplot

sns.histplot(data=iris , x='Sepal.Length' , bins=16 , hue='Species')
plt.show()
```

```python
# densityplot: sns.kdeplot

sns.kdeplot(data=iris, x='Sepal.Length', hue='Species')
plt.show()
```

```python
# boxplot

sns.boxplot(data=iris, y='Sepal.Length', x='Species')
plt.show()
```

### distplot : histogram + density plot

```python
# titanic['Fare'] 에 대해서 distplot 그려보기

sns.distplot(titanic['Fare'], bins=30, hist_kws={'edgecolor':'gray'})
plt.show()
```

```python
# boston['medv'] 에 대해서 distplot을 그려보기

sns.distplot(boston['medv'], bins=30, hist_kws={'edgecolor':'gray'})
plt.show()
```

### jointplot : scatter + histogram

```python
# titanic['Age'] 와 titanic['Fare']에 대해 그려보기

sns.jointplot(x='Age', y='Fare', data=titanic)
plt.show()
```

```python
# 위 내용에 생존여부 별로 구분해서 그려보기

sns.jointplot(x='Age', y='Fare', data=titanic, hue='Sex')
plt.show()
```

### pairplot : scatter + histogram (확장)

```python
# air 데이터셋에 대해 pairplot 그려보기

sns.pairplot(air)
plt.show()
```

```python
# 월별로 구분해서 그려보기

sns.pairplot(air, hue='Month')
plt.show()
```

### countplot : 집계 + bar plot

```python
# boston['chas']에 대해 그려보기

sns.countplot(boston['chas'])
plt.show()
```

```python
# titanic['Survived']에 대해 성별 별로 구분해서 그려보기
sns.countplot(x = titanic['Survived'], hue='Sex', data=titanic)
plt.show()
```

### barplot : 평균비교 bar plot + error bar

* 범주별 (x), 숫자(y)의 평균을 비교하는 그래프이다

```python
# boston['chas']별 boston['medv']의 평균을 비교

sns.barplot(x=boston['chas'], y=boston['medv'])
plt.show()
```

### heatmap : 두 범주 집계 시각화

* 집계와 피봇을 먼저 만들어줘야함

```python
# titanic 데이터에서 Embarked별 Pclass별 passengerID count 값을 구하시오 (temp 변수 선언)

temp = titanic.groupby(['Embarked','Pclass'], as_index = False)['PassengerId'].count()
temp1 = temp.pivot('Embarked','Pclass','PassengerId')

print(temp1)
```

```python
sns.heatmap(temp1, annot = True)
plt.show()
```
